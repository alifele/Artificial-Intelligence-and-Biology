```python
import numpy as np
import matplotlib.pyplot as plt
from network import Neural_Network
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_shape = 1
hidden_shape = [4,2,4,3]
output_shape = 1
```

## Generating the data set

```python
x = np.linspace(0,10,1000)
y = np.sin(x) + 0.1*np.random.random(x.shape)
plt.plot(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x.reshape((-1,1)),y.reshape((-1,1)), shuffle = True)
```

```python
plt.plot(x_train, y_train,'8',label = 'training data')
plt.plot(x_test, y_test, '8', label = 'test data')
plt.legend()
```

## Training the model

```python
model = Neural_Network(input_shape, hidden_shape, output_shape)
```

```python
model.fit(x_train, y_train, epochs = 200, batch_size=512, alpha = 0.0005)
```

```python
plt.figure(figsize=(13,4))
plt.plot(model.cost_list,'--', lw = 3)
plt.title('loss over epochs', size=20)
plt.grid('on')
plt.xlabel('epochs', size = 13)
plt.ylabel('error', size = 13)
```
